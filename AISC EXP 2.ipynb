# Activation Functions

```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
X=[0.2*i for i in range(-50,51)]
X=np.array(X)
```

## Identity Function

```python
def identity(x):
    return x
```

```python
ax = plt.gca()
Y=identity(X)
ax.plot(X, Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

plt.show()
```

## Binary Step

```python
def bin_step(x):
    threshold = 0 
    x[x>=threshold]=1
    x[x<threshold]=0
    return x
```

```python
Y=bin_step(X)
```

```python
X=[0.2*i for i in range(-50,51)]
X=np.array(X)
```

```python
ax = plt.gca()

ax.plot(X, Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

plt.show()
```

## Bipolar step 

```python
X=[0.2*i for i in range(-50,51)]
X=np.array(X)
```

```python
def bip_step(x):
    threshold = 0 
    x[x>=threshold]=1
    x[x<threshold]=-1
    return x
```

```python
Y=bip_step(X)
```

```python
X=[0.2*i for i in range(-50,51)]
X=np.array(X)
```

```python
ax = plt.gca()

ax.plot(X, Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.show()
```

## Sigmoid

```python
def bin_sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
ax = plt.gca()

ax.plot(X,bin_sigmoid(X))
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.show()
```

## Tansig Function

```python
def tansig(x):
    return 2/(1+np.exp(-2*x))-1
```

```python
ax = plt.gca()

ax.plot(X,tansig(X))
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.show()
```

## Satuarated Linear

```python
def lin_sat(x):
    x[x>2.5]=2.5
    x[x<-2.5]=-2.5
    return x
```

```python
ax = plt.gca()
Y=lin_sat(X)
X=[0.2*i for i in range(-50,51)]
X=np.array(X)
ax.plot(X,Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.show()
```

## ReLu Function

```python
def relu(x):
    x[x<0]=0
    return x
```

```python
Y=relu(X)
```

```python
X=[0.2*i for i in range(-50,51)]
X=np.array(X)
```

```python
ax = plt.gca()

ax.plot(X, Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.show()
```

## Leaky Relu

```python
def l_relu(x):
    x[x<0]=0.1*x[x<0]
    return x
```

```python
Y=l_relu(X)
```

```python
X=[0.2*i for i in range(-50,51)]
X=np.array(X)
```

```python
ax = plt.gca()

ax.plot(X, Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.show()
```

## SoftPlus

```python
def softplus(x):
    return np.log(1+np.exp(x))
```

```python
Y=softplus(X)
```

```python
ax = plt.gca()

ax.plot(X, Y)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.show()
```

```python
def bin_sigmoid(x):
    return 1/(1+np.exp(-(x+0.5)))-1/(1+np.exp(-(x-.5)))
```

```python
ax = plt.gca()

ax.plot(X,bin_sigmoid(X))
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')

plt.show()
```
